## Ejemplo de como recuperar la secuencia de los distintos exones

```python
# Cargamos en el Notebook las secuencias de los exones en formato .fasta

sec_exones_ctnnb1 = SeqIO.parse ("data_CTNNB1/exones_ctnnb1.fasta", "fasta")
```

```python
# Obtenemos la secuencia del primer exón

secuencia1 = next (sec_exones_ctnnb1)

print (secuencia1)
```

```python
# Obtenemos la secuencia del segundo exón

secuencia2 = next (sec_exones_ctnnb1)

print (secuencia2)
```

```python
# Obtenemos la secuencia de un exón a partir de su id

for i in SeqIO.parse ("data_CTNNB1/exones_ctnnb1.fasta", "fasta"):
    if i.id == "ENSE00003593193":
        secuencia3 = i
        
print (secuencia3)
```
