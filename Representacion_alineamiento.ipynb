```python
from Bio import AlignIO
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Vamos a cargar los datos que tenemos del alineamiento en el Notebook

```python
alineamiento = AlignIO.read ("data_TP53/alineamiento_var_patogenicas_exon4_tp53.aln-clustal_num", "clustal")
```

```python
print (alineamiento)
```

```python
def alignment_to_dataframe(alignment):
    records = list(alignment)
    data = []
    for record in records:
        row = list(record.seq)
        data.append(row)
    df = pd.DataFrame(data, index=[rec.id for rec in records])
    return df

df_alignment = alignment_to_dataframe(alineamiento)

# Crear un mapa de calor del alineamiento
plt.figure(figsize=(200, 200))
sns.heatmap(df_alignment.apply(lambda x: [ord(i) for i in x]), cmap="viridis", cbar=False, xticklabels=True, yticklabels=True)
plt.title("Heatmap del Alineamiento de Variantes Patogénicas del Gen TP53")
plt.xlabel("Posición en la Secuencia")
plt.ylabel("ID de la Secuencia")
plt.show()

plt.savefig('heatmap.png',  dpi=300, bbox_inches='tight')

# Mostrar el heatmap en pantalla
plt.show()
```
